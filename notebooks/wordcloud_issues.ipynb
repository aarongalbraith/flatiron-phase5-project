For a simple comparison of different groups of documents, the individuality of the records is not important. All records of a common type are joined together before they are tokenized.

1. Extract documents from a dataframe as a list of strings: `documents = df[df.col = label].text.to_list()`
2. Combine the documents into a single string: `joined_docs = ' '.join(documents)`
3. Tokenize this into a list of tokens: `tokens = nltk.word_tokenizer(joined_docs)`
4. Join these tokens into a single string: `joined_tokens = ','.join(tokens)`

In modeling, where we wish to preserve attributes of the individual records, each record is tokenized.
